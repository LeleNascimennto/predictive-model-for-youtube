```python
def ler_e_exibir_html_historico():
    base_dir = Path("../../Takeout/YouTube e YouTube Music/histórico")
    arquivos_html = list(base_dir.glob("*.html"))

    print(f"Total de arquivos HTML encontrados: {len(arquivos_html)}\n")

    for arquivo in arquivos_html:
        print(f"\nArquivo: {arquivo.name}")

        with open(arquivo, "r", encoding="utf-8") as f:
            soup = BeautifulSoup(f, "html.parser")

        dados = []

        blocos = soup.find_all(
            "div",
            class_="content-cell mdl-cell mdl-cell--6-col mdl-typography--body-1"
        )

        for bloco in blocos:
            link = bloco.find("a")
            if not link:
                continue

            # termo pesquisado
            titulo = link.text.strip()
            url = link["href"]

            # texto completo do bloco
            texto = bloco.get_text(separator=" ", strip=True)

            # tipo da ação (antes do link)
            tipo = texto.split(titulo)[0].strip()

            # data (depois do link)
            data = texto.split(titulo)[-1].strip()

            dados.append([tipo, titulo, url, data])

        df = pd.DataFrame(dados, columns=["tipo", "titulo", "url", "data"])

        print("\n--- dtypes antes de convert_dtypes() ---")
        display(df.dtypes)

        df = df.convert_dtypes()

        print("\n--- dtypes após convert_dtypes() ---")
        display(df.dtypes)

        print(f"\n📊 Dados (primeiras 20 linhas):")
        display(df.head(20))
```

```python
ler_e_exibir_html_historico()
```
